# A-1 & A-2

```python
class Employee:
    def __init__(self,name,number,address,salary,post,year_of_joining,education,year_of_experience):
        self.name=name
        self.number=number
        self.address=address
        self.salary=salary
        self.post=post
        self.year_of_joining=year_of_joining
        self.education=education
        self.year_of_experience=year_of_experience

    def describe_employee(self):
        return '{} {} {} {} {} {} {} {} {} {} {} {} {} {} {} {} {} {} {} {} {} {} {}'.format('Employee Name: ',self.name,'\n','Contact Info.',self.number,'\n',
                                           'Address Detail:',self.address,'\n','Salary:',self.salary,'\n','Post:',self.post,'\n',
                                            'Year of Joining:',self.year_of_joining,'\n','Education:',self.education,'\n',
                                              'Year Of Experience:',self.year_of_experience  )

    def update_year_of_experience(self,update_experience):
    
        update_experience=self.year_of_experience
        return '{} {}'.format('Update Year of experience is',update_experience)
    def increase_salary(self,update_salary):
        up_sa=self.salary+update_salary
        return '{} {}'.format('Updated Salary is ',up_sa)
        




        
```

```python

emp101=Employee('Ali Ahmed',923142345674,'karachi,Sindh',75000,'Marketing head',2012,'BS Commerce',5)

print('*** Calling Indivdual Attribute ***')
print(emp101.name)
print(emp101.number)
print(emp101.address)
print(emp101.salary)

print('\n''*** Complete Detail of an Employee: ***')
print(emp101.describe_employee())

print('\n''*** Updating Employee Year of Experience: ***')
print(emp101.update_year_of_experience(7))

print('\n''*** Updating Salary: ***')
print(emp101.update_salary(100000))
```
